4.Удалить из набора поля, не несущие информативной нагрузки - содержащие уникальные либо константные значения

```python
df = df[df.columns[df.nunique() != df.shape[0]]]
df = df[df.columns[df.nunique() >1]]
print(df)
```

5.Заменить возраст пассажиров на год рождения

```python
df['Age'] = 2022 - df['Age']
df
```

6.Переименовать все названия колонок на русском языке

```python
df = df.rename(columns={'PassengerId':'ID_пассажира', 'Survived':'Прибыл', 'Pclass':'Пассажирский_класс', 'Name':'ФИО', 'Sex':'Пол',
                       'Age':'Год рождения', 'SibSp':'Кол-во родственников на борту','Parch':'Жажда', 'Ticket':'Билет', 'Fare':'Стоимость проезда', 'Cabin':'Место', 'Embarked':'Посадка'})
df
```

7.Заменить все отсутствующие данные средним значением (в случае числового поля) либо самым распространенным в категориальных полях.

```python
df['Прибыл'] = df['Прибыл'].fillna(round(df['Прибыл'].mean(),0))
df['Пассажирский_класс'] = df['Пассажирский_класс'].fillna(round(df['Пассажирский_класс'].mean(),0))
df['Год рождения'] = df['Год рождения'].fillna(round(df['Год рождения'].mean(),0))
df['Кол-во родственников на борту'] = df['Кол-во родственников на борту'].fillna(round(df['Кол-во родственников на борту'].mean(),0))
df['Жажда'] = df['Жажда'].fillna(round(df['Жажда'].mean(),0))
df['Стоимость проезда'] = df['Стоимость проезда'].fillna(round(df['Стоимость проезда'].mean(),0))
df['Пол'] = df['Пол'].fillna(df['Пол'].mode()[len(df['Пол'].mode())-1])
df['Место'] = df['Место'].fillna(df['Место'].mode()[len(df['Место'].mode())-1])
df['Билет'] = df['Билет'].fillna(df['Билет'].mode()[len(df['Билет'].mode())-1])
df['Посадка'] = df['Посадка'].fillna(df['Посадка'].mode()[len(df['Посадка'].mode())-1])


df
```
